```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
image_dir=r"C:\Users\bodla\OneDrive - Mahindra University\Pictures\Camera Roll\WIN_20220902_22_02_52_Pro.jpg"

```

```python
from skimage.io import imread,imshow
image_ski=imread(image_dir,as_gray=False)
imshow(image_ski)
```

```python
r,g,b=image_ski[:,:,0],image_ski[:,:,1],image_ski[:,:,2]
r=256-r
g=256-g 
b=256-b
neg_img=r+g+b
imshow(neg_img)
```

```python
from numpy import fliplr,flipud
lr_img=fliplr(image_ski)
imshow(lr_img)
```

```python
ud_img=flipud(image_ski)
imshow(ud_img)
```

```python
ud_lr_img=fliplr(ud_img)
imshow(ud_lr_img)
```

```python
#image thresholding
import cv2 as cv

ret,thresh1 = cv.threshold(image_ski,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(image_ski,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(image_ski,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(image_ski,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(image_ski,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [image_ski, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    # plt.xticks([]),plt.yticks([])
plt.show()
```

```python
from skimage import exposure
image_1=exposure.adjust_gamma(image_ski,gamma=0.5,gain=1)
image_2=exposure.adjust_gamma(image_ski,gamma=2,gain=1)
image_3=exposure.adjust_gamma(image_ski,gamma=4,gain=1)
image_4=exposure.adjust_gamma(image_ski,gamma=6,gain=1)
image_5=exposure.adjust_gamma(image_ski,gamma=8,gain=1)
```
